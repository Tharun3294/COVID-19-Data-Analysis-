```python
import seaborn as sns
import matplotlib.pyplot as plt
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
df['Month'] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Total_Cases'], bins=30, kde=True, color='blue')

plt.title("Distribution of Confirmed COVID-19 Cases", fontsize=14)
plt.xlabel("Confirmed Cases", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="region", y="Total_Deaths", data=df)
plt.xticks(rotation=90)
plt.title("Deaths Distribution by region")
plt.show()
```

```python
sns.scatterplot(x="Total_Cases", y="Total_Deaths",hue="region", data=df)
plt.title("COVID-19 Cases vs. Deaths")
plt.show()
```

```python
sns.barplot(x="country", y="Total_Cases", data=df)
plt.title("Total Confirmed Cases by region")
plt.xlabel("country")
plt.ylabel("Total_Cases")
```

```python
sns.barplot(x="country", y="Total_Deaths", data=df)
plt.title("Total Deaths Cases by region")
plt.xlabel("country")
plt.ylabel("Total_Deaths")
```

```python
sns.barplot(x="country", y="Total_Recovered", data=df)
plt.title("Total Recovered Cases by region")
plt.xlabel("country")
plt.ylabel("Total_Recovered")
```

```python
plt.figure(figsize=(12, 6))
df_grouped = df.groupby('Date')['Total_Cases'].sum().reset_index()
sns.lineplot(x='Date', y='Total_Cases', data=df_grouped, marker='o', color='b')

plt.title("Trend of COVID-19 Cases Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Confirmed Cases", fontsize=12)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df_monthly = df.groupby('Month')['Total_Cases'].sum().reset_index()

sns.barplot(x='Month', y='Total_Cases', data=df_monthly, palette='coolwarm')
plt.title("Monthly COVID-19 Cases", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Cases", fontsize=12)
plt.xticks(range(0, 13))
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df_yearly = df.groupby(['Year', 'country'])['Total_Cases'].sum().reset_index()

sns.barplot(x='Year', y='Total_Cases', hue='country', data=df_yearly)
plt.title("Yearly COVID-19 Cases per Country", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Cases", fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()

```

```python
plt.figure(figsize=(12, 6))

df_grouped = df.groupby('Date')[['Total_Cases', 'Total_Deaths', 'Total_Recovered']].sum().reset_index()

sns.lineplot(x='Date', y='Total_Cases', data=df_grouped, label='Total_Cases', color='blue')
sns.lineplot(x='Date', y='Total_Deaths', data=df_grouped, label='Total_Deaths', color='red')
sns.lineplot(x='Date', y='Total_Recovered', data=df_grouped, label='Total_Recovered', color='green')

plt.title("COVID-19 Trends Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Cases Count", fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

df_monthly = df.groupby('Month')[['Total_Cases', 'Total_Deaths', 'Total_Recovered']].sum().reset_index()

df_monthly.plot(x='Month', kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

plt.title("Monthly COVID-19 Cases, Deaths, and Recoveries", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Count", fontsize=12)
plt.legend(["Total_Cases", "Total_Deaths", "Total_Recovered"])
plt.show()
```

```python
plt.figure(figsize=(14, 6))

df_country = df.groupby('region')[['Total_Cases', 'Total_Deaths', 'Total_Recovered']].sum().reset_index()
df_country = df_country.sort_values(by='Total_Cases', ascending=False).head(10)  # Top 10 countries

df_country.plot(x='region', kind='bar', stacked=True, figsize=(14, 6), colormap='plasma')

plt.title("COVID-19 Cases, Deaths, and Recoveries by Country", fontsize=14)
plt.xlabel("country", fontsize=12)
plt.ylabel("Total Count", fontsize=12)
plt.legend(["Total_Cases", "Total_Deaths", "Total_Recovered"])
plt.show()
```

```python
#                            DATA VISULIZATION FOR INDIA
df_india = df[df['region'] == 'India']
india_summary = df_india.groupby('Date')[['Total_Cases', 'Total_Deaths', 'Total_Recovered']].sum().reset_index()
print(india_summary.tail())  # Check last few records
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Total_Cases', data=india_summary, label='Total_Cases', color='blue')
sns.lineplot(x='Date', y='Total_Deaths', data=india_summary, label='Total_Deaths', color='red')
sns.lineplot(x='Date', y='Total_Recovered', data=india_summary, label='Total_Recovered', color='green')
plt.title("COVID-19 Trend in India", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Cases", fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

df_state = df_india.groupby('Province')[['Total_Cases', 'Total_Deaths', 'Total_Recovered']].sum().reset_index()
df_state = df_state.sort_values(by='Total_Cases', ascending=False)

df_state.plot(x='Province', kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm')

plt.title("COVID-19 Cases, Deaths, and Recoveries by State in India", fontsize=14)
plt.xlabel("Province", fontsize=12)
plt.ylabel("Total Count", fontsize=12)
plt.xticks(rotation=90)
plt.legend(["Total_Cases", "Total_Deaths", "Total_Recovered"])
plt.show()
```
